```python
from google.colab import drive
drive.mount('/content/drive')

folder_path = '/content/drive/My Drive/directory/dataset'
```

```python
import os
from rdflib import Graph
from shexer.shaper import Shaper
from shexer.consts import TURTLE, SHEXC

namespaces_dict = {
    "http://example.org/": "ex",
    "http://www.cidoc-crm.org/cidoc-crm/": "crm",
    "http://weso.es/shapes/": "",
    "http://www.w3.org/2001/XMLSchema#": "xsd"
}

combined_graph = Graph()
for filename in os.listdir(folder_path):
    if filename.endswith('.ttl'):
        file_path = os.path.join(folder_path, filename)
        print(f"Parsing {file_path}")
        combined_graph.parse(file_path, format="turtle")

# Convert the combined graph to a string in Turtle format
raw_graph = combined_graph.serialize(format="turtle")



# Initialize the Shaper with the combined graph
shaper = Shaper(
    all_classes_mode =True,
    raw_graph=raw_graph,
    input_format=TURTLE,
    namespaces_dict=namespaces_dict,
    instantiation_property="http://www.w3.org/1999/02/22-rdf-syntax-ns#type"
)


# Specify the output file path
output_file = "/content/drive/My Drive/directory/ShapeDataset.shex"

# Generate the ShEx graph
shaper.shex_graph(
    output_file=output_file,
    acceptance_threshold=0  # Adjust this threshold as needed
)

print(f"Done! Shape expressions saved to {output_file}")
```
